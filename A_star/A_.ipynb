```python
#Bai tap vn
import heapq

# Tính khoảng cách Manhattan (có thể thay bằng Euclidean)
def distance(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

# Hàm heuristic: khoảng cách nhỏ nhất đến điểm chưa giao
def heuristic(current, remaining):
    if not remaining:
        return 0
    return min(distance(current, p) for p in remaining)

def a_star_delivery(start, deliveries):
    # Trạng thái: (f, g, current_position, remaining_points, path)
    initial_state = (0, 0, start, deliveries, [start])
    frontier = []
    heapq.heappush(frontier, initial_state)
    visited = set()

    while frontier:
        f, g, current, remaining, path = heapq.heappop(frontier)

        # Nếu đã giao hết
        if not remaining:
            return path, g

        state_key = (current, tuple(sorted(remaining)))
        if state_key in visited:
            continue
        visited.add(state_key)

        # Mở rộng các trạng thái kế tiếp
        for next_point in remaining:
            new_remaining = [p for p in remaining if p != next_point]
            new_g = g + distance(current, next_point)
            new_h = heuristic(next_point, new_remaining)
            new_f = new_g + new_h
            new_path = path + [next_point]
            heapq.heappush(frontier, (new_f, new_g, next_point, new_remaining, new_path))

    return None, float("inf")

# Ví dụ chạy
start = (0, 0)
deliveries = [ (3, 1), (1, 2), (4, 4)]
path, cost = a_star_delivery(start, deliveries)

print("Đường đi tối ưu:", path)
print("Tổng chi phí:", cost)
```
